### import libs

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url = "https://meteostat.net/en/place/vn/thu-uc?s=48900&t=2024-05-19/2024-05-26"
```

Beautiful Soup

```python
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
wait = WebDriverWait(driver, 10)
model = wait.until(EC.presence_of_element_located((By.ID, 'cookieModal')))
model.find_element(By.CLASS_NAME, 'modal-footer').find_elements(By.TAG_NAME, 'button')[1].click()
```

```python
model = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'btn-group')))
model.find_elements(By.TAG_NAME,'button')[2].click()
```

```python
for i in range(10):
    model = wait.until(EC.presence_of_element_located((By.ID, 'details-modal')))
    model.find_elements(By.TAG_NAME,'button')[1].click()
```

```python
model = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'table-responsive')))
table = model.find_element(By.TAG_NAME,'table')
table = table.find_element(By.TAG_NAME,'tbody')
rows = table.find_elements(By.TAG_NAME, 'tr')
data = []
for row in rows:
    temp = []
    cols = row.find_elements(By.TAG_NAME, 'th')
    for ele in cols:
        temp.append(ele.text.strip())
    cols = row.find_elements(By.TAG_NAME, 'td')
    for ele in cols:
        temp.append(ele.text.strip())

    data.append(temp)

```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("weather.csv", index=False)
```
